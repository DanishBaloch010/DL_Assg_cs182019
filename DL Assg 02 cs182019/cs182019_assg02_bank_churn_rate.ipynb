```python
import pandas as pd

dataframe = pd.read_csv('Churn_Modelling.csv')
```

## dropping rownumber, customerID and surname(first three columns) , as they do not add meaning to the data.

```python
dataframe.drop(['RowNumber', 'CustomerId', 'Surname'], axis = 1, inplace=True)
```

```python
dataframe
```

```python
def get_unique_values(df):
    for col in df.columns:
        print(col)
        unique_val = df[col].unique()
        print(unique_val)
        print(col , 'total unique values:',len(unique_val))
        print()
```

```python
get_unique_values(dataframe)
```

## checking if estimatedSalary have zero entry or not ? because if estimatedSalary is zero than the respective record is invalidated. 

```python
def check_zero_entry_on_float_column(col):
    unique_items = dataframe[col].unique()
    print(unique_items)
    print(len(unique_items))

    check = False
    for idx , val in enumerate(unique_items):
        if val == 0.0:
            check = True
            print('There is zero entry at index as follows:')
            print(idx)
            print(val)
    if check == False:
        print('There is no Zero Entry in column :' , col)
```

```python
check_zero_entry_on_float_column('EstimatedSalary')
```

##  balance feature can have 0 entry , in this case it makes sense because every salary feature is filled. 

```python
check_zero_entry_on_float_column('Balance')
```

## linear scaling CreditScore , Age and Balance and estimatedSalary Column.

```python
dfscaled = dataframe.copy()
columns = ['CreditScore' , 'Age' , 'Balance' , 'EstimatedSalary']
for col in columns:
    dfscaled[col] = (dfscaled[col] - dfscaled[col].min()) / (dfscaled[col].max() - dfscaled[col].min())    
    
# lineartly normalized
display(dfscaled)
```

# converting categorical values to quantitative values

### as we know there are only 3 unique values in geography column hence,
### 0 - > France , 1 - > Spain, 2 - > Germany

### as we know there are only 2 unique values in gender column hence,
### 0 - > Male  ,1 - > Female

```python
for idx ,val in dfscaled.iterrows():
    if  dfscaled.loc[idx,'Gender'] == 'Male':
        dfscaled.loc[idx,'Gender'] = '0'
    elif dfscaled.loc[idx,'Gender'] == 'Female':
        dfscaled.loc[idx,'Gender'] = '1'
        
for idx ,val in dfscaled.iterrows():
    if  dfscaled.loc[idx,'Geography'] == 'France':
        dfscaled.loc[idx,'Geography'] = '0'
    elif dfscaled.loc[idx,'Geography'] == 'Spain':
        dfscaled.loc[idx,'Geography'] = '1'
    elif dfscaled.loc[idx,'Geography'] == 'Germany':
        dfscaled.loc[idx,'Geography'] = '2'
```

```python
dfscaled
```

## applying softmax on NumOfProducts , as it will reduce the scale to 1. 
### unique values are 1 , 2 , 3 , 4 

```python
import math
sum_of_exp = 0
for i in range(1,5):
    sum_of_exp =  sum_of_exp + math.exp(i)
    
print(sum_of_exp)

for idx ,val in dfscaled.iterrows():
    if  dfscaled.loc[idx,'NumOfProducts'] == 1:
        dfscaled.loc[idx,'NumOfProducts'] = (math.exp(1)/(sum_of_exp))
    elif dfscaled.loc[idx,'NumOfProducts'] == 2:
        dfscaled.loc[idx,'NumOfProducts'] = (math.exp(2)/(sum_of_exp))
    elif  dfscaled.loc[idx,'NumOfProducts'] == 3:
        dfscaled.loc[idx,'NumOfProducts'] = (math.exp(3)/(sum_of_exp))
    elif dfscaled.loc[idx,'NumOfProducts'] == 4:
        dfscaled.loc[idx,'NumOfProducts'] = (math.exp(4)/(sum_of_exp))
```

## changing objects to int64

```python
import numpy as np
dfscaled['Geography'] = dfscaled['Geography'].astype(str).astype(np.int64)
dfscaled['Gender'] = dfscaled['Gender'].astype(str).astype(np.int64)
```

```python
result = dfscaled.dtypes
print(result)
```

```python
X = dfscaled.drop('Exited', axis = 1)
Y = dfscaled['Exited']
```

```python
X = np.asarray(X)
Y = np.asarray(Y)

print(X)
print(Y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.33, random_state=42)
print(type(X_train))
print(type(Y_train))
print(type(X_test))
print(type(Y_test))
```

```python
print(len(X_train))
print(len(X_train[0]))
print(len(Y_train))
print(Y_train)

print('-------------------------\n')

print(len(X_test))
print(len(X_test[0]))
print(len(Y_test))
print(Y_test)
```

```python
import keras
def model(x_train, x_test, y_train, y_test):
    mod = keras.Sequential([
        keras.layers.Dense(1500, input_dim = len(x_train[0]), activation = 'relu'),
        keras.layers.Dense(700, activation = 'relu'),
        keras.layers.Dense(1, activation = 'sigmoid'),
    ])
    
    mod.compile(
        optimizer='adam',
        loss='binary_crossentropy',
        metrics = ['accuracy']
    )
    
    mod.fit(x_train, y_train, epochs=110)
    
    print(f'\n test accuracy -> {mod.evaluate(x_test, y_test)}')

    y_pred = mod.predict(x_test)
    y_pred = np.round(y_pred)
    
    return y_pred
```

```python
y_pred = model(X_train, X_test, Y_train, Y_test)
```

```python
print(y_pred)
```

## classification report

```python
#Evaluating the prediction
from sklearn.metrics import confusion_matrix , classification_report

print(classification_report(Y_test,y_pred))
```
