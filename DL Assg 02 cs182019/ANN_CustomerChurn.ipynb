```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
%matplotlib inline
```

```python
df = pd.read_csv("customer_churn.csv")
df.sample(5)
```

```python
df.drop('customerID',axis='columns',inplace=True)
```

```python
df[pd.to_numeric(df.TotalCharges,errors='coerce').isnull()]
```

```python
df1 = df[df.TotalCharges!=' ']
df1.shape
#remove space in total charges column
```

```python
#Store float values in df1 dataframe
df1.TotalCharges = pd.to_numeric(df1.TotalCharges)
```

```python
#check the values
df1.TotalCharges.values
```

```python
#just confirming its data type
df1.TotalCharges.dtypes
```

```python
# Creating a function to iterate through columns that consist of logical values:
def print_unique_col_values(df):
  for column in df:
    if df[column].dtypes=='object':
      print(f'{column}: {df[column].unique()}') 
```

```python
print_unique_col_values(df1)
```

```python
#Some data cleaning to ensure that all values have proper options
df1.replace('No internet service','No',inplace=True)
df1.replace('No phone service','No',inplace=True)
df1['gender'].replace({'Female':1,'Male':0},inplace=True)
```

```python
print_unique_col_values(df1)
```

```python
#Converting all values into 0 and 1
yes_no_columns = ['Partner','Dependents','PhoneService','MultipleLines','OnlineSecurity','OnlineBackup',
                  'DeviceProtection','TechSupport','StreamingTV','StreamingMovies','PaperlessBilling','Churn']
for col in yes_no_columns:
  df1[col].replace({'Yes': 1,'No': 0},inplace=True)
```

```python
for col in df1:
  print(f'{col}: {df1[col].unique()}')
```

```python
#One hot encoding other columns
df2 = pd.get_dummies(data=df1, columns=['InternetService','Contract','PaymentMethod'])
df2.columns
```

```python
df2.dtypes
```

```python
#In Deep Learning, it is imp to scale the data to ensure that it is between some range
df2.sample(5)
```

```python
cols_to_scale = ['tenure','MonthlyCharges','TotalCharges']

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df2[cols_to_scale] = scaler.fit_transform(df2[cols_to_scale])
```

```python
df2.sample(5)
```

```python
#train test split for our ANN
X = df2.drop('Churn',axis='columns')
y = df2['Churn']

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=5)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
X_train[:10]
```

```python
len(X_train.columns)
```

```python
import tensorflow as tf
from tensorflow import keras


#model = keras.Sequential([
   # keras.layers.Dense(26, input_shape=(26,)), #input layer
  #  keras.layers.Dense(15, activation='relu'), #hidden layer
 #   keras.layers.Dense(1, activation='sigmoid') #ouput layer
#])

#or this will also work
model = keras.Sequential([
    keras.layers.Dense(26, input_shape=(26,), activation='relu'),
    keras.layers.Dense(1, activation='sigmoid') #ouput layer
])

model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])

model.fit(X_train, y_train, epochs=50)
```

```python
yp = model.predict(X_test)
```

```python
#Checking the first few values of prediction
yp[:5]
```

```python
#Convert probabilistic values to 0,1
y_pred = []
for element in yp:
    if element > 0.5:
        y_pred.append(1)
    else:
        y_pred.append(0)
```

```python
y_pred[:5]
```

```python
#Evaluating the prediction
from sklearn.metrics import confusion_matrix , classification_report

print(classification_report(y_test,y_pred))
```
